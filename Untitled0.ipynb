```python
# Imports
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

# =========================
# Load SMALL subset (FAST)
# =========================
# Use header=None because the CSV does not have a header row
df = pd.read_csv('/content/sample_data/mnist_train_small.csv', header=None)

# The first column is the label, and the rest are pixel values
df.rename(columns={0: 'label'}, inplace=True)

df = df.sample(5000, random_state=42)

X = df.drop('label', axis=1).values / 255.0
y = df['label'].values

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

# Convert to tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)
X_val = torch.tensor(X_val, dtype=torch.float32)
y_val = torch.tensor(y_val, dtype=torch.long)

train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=64, shuffle=True)

# =========================
# Model
# =========================
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 64)
        self.fc2 = nn.Linear(64, 10)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

model = Net()

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# =========================
# Training + Metrics
# =========================
epochs = 5

train_losses = []
val_losses = []
train_accs = []
val_accs = []
grad_norms = []
weight_norms = []

for epoch in range(epochs):
    model.train()
    running_loss = 0
    correct = 0
    total = 0
    grad_total = 0

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()

        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)

        loss.backward()

        # Gradient norm (for graph)
        total_norm = 0
        for p in model.parameters():
            if p.grad is not None:
                total_norm += p.grad.data.norm(2).item()
        grad_total += total_norm

        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

    # Training metrics
    train_loss = running_loss / len(train_loader)
    train_acc = 100 * correct / total

    # Validation
    model.eval()
    with torch.no_grad():
        val_outputs = model(X_val)
        val_loss = criterion(val_outputs, y_val).item()
        _, val_pred = torch.max(val_outputs, 1)
        val_acc = 100 * (val_pred == y_val).sum().item() / len(y_val)

    # Weight norm
    weight_norm = sum(p.data.norm(2).item() for p in model.parameters())

    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    grad_norms.append(grad_total)
    weight_norms.append(weight_norm)

    print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Val Acc={val_acc:.2f}%")

# =========================
# 📊 PLOTS (6 GRAPHS)
# =========================

# 1. Train Loss
plt.figure()
plt.plot(train_losses)
plt.title("Train Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# 2. Validation Loss
plt.figure()
plt.plot(val_losses)
plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# 3. Train Accuracy
plt.figure()
plt.plot(train_accs)
plt.title("Train Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.show()

# 4. Validation Accuracy
plt.figure()
plt.plot(val_accs)
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.show()

# 5. Gradient Norm
plt.figure()
plt.plot(grad_norms)
plt.title("Gradient Norm")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.show()

# 6. Weight Norm
plt.figure()
plt.plot(weight_norms)
plt.title("Weight Norm")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.show()
```
